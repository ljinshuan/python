```python
from scipy.spatial.distance import cosine

def calCos(a,b):
    return 0.5+0.5*(1-cosine(a, b))
```

```python
user_like_matrix=rating_matrix.corr(method=calCos)
```

```python
def calLikeRate(userId,movieId):
    # 取最相近的用户 与 评价过movieId的用户的交集
    users=pd.Series( list( set(like_users.index).intersection(set(user_rating_actions[movieId][user_rating_actions[movieId].notnull()].index))))
    # 计算偏好
    return users.map(lambda d: user_like_matrix[userId][d]).sum()
```

```python
data=pd.Series( user_rating_actions.loc[1].index,index=user_rating_actions.loc[1].index)
result=data.map(lambda d:calLikeRate(1,d))
```
